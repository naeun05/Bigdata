```python
page = 0
review_content = []

url = "https://www.tripadvisor.co.kr/Attraction_Review-g1143545-d2224863-Reviews-Jeonju_Hanok_Village-Jeonju_Jeollabuk_do.html"
driver = webdriver.Chrome()
driver.get(url)
time.sleep(4)

while True:
    page +=1
    more_btn = driver.find_element_by_class_name("XUVJZtom")
    more_btn.click()
    reviews = driver.find_elements_by_class_name("IRsGHoPm")
    for i in range(len(reviews)):
        review = reviews[i].text
        review = review.replace("\n",' ')
        review_content.append(review)
    try:
        next_btn = driver.find_element_by_css_selector("a.ui_button.nav.next.primary ")
        next_btn.click()
    except:
        print("완료")
        break
        
    time.sleep(4)
    print(page)
```

```python
data = pd.DataFrame({"review_content":review_content})
data.to_csv("data/전라북도1_전주한옥마을.csv", encoding="utf-8-sig")

f = open('data/전라북도1_전주한옥마을.txt', 'w', encoding='utf-8')
 
for r in review_content:  
    f.write(r)
    f.write('\n')
```
